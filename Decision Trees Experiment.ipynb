```python
#Decision Trees Experiment
#Determine whether a mushroom is edible
## Source: Edx 6.5 Lab ##

#Setup

import pandas as pd

#https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.names

#Load the mushroom dataset

X = pd.read_csv("agaricus-lepiota.data", na_values = '?', names = ["Class",
                "Cap-Shape", "Cap-Surface", "Cap-Color", "Bruises", "Odor", "Gill-Attach", 
                "Gill-Spacing", "Gill-Size", "Gill-Color", "Stalk-Shape", "Stalk-Root",
                "Stalk-Surface-Above", "Stalk-Surface-Below", "Stalk-Color-Above",
                "Stalk-Color-Below", "Veil-Type", "Veil-Color", "Ring-Number", "Ring-Type",
                "Spore-Color", "Population", "Habitat"])
X.head()
```

```python
# drop NaNs

X = X.dropna(axis = 0)
```

```python
# Pull labels out of dataset

Y = X[["Class"]]
X = X.drop(labels = "Class", axis = 1)
Classes = ['e', 'p']
Y.Class = Y.Class.astype("category", categories = Classes).cat.codes
Y.head()
```

```python
# Code the categorical features into numeric values

columns = ["Cap-Shape", "Cap-Surface", "Cap-Color", "Bruises", "Odor", "Gill-Attach", "Gill-Spacing", "Gill-Size", 
 "Gill-Color", "Stalk-Shape", "Stalk-Root", "Stalk-Surface-Above", "Stalk-Surface-Below", "Stalk-Color-Above",
 "Stalk-Color-Below", "Veil-Type", "Veil-Color", "Ring-Number", "Ring-Type","Spore-Color", "Population", "Habitat"]
for col in columns:
    X[col] = X[col].astype("category").cat.codes
X.head()
```

```python
#Split train/test sets

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3, random_state = 7)
```

```python
#Import and fit DT model

from sklearn import tree
model = tree.DecisionTreeClassifier()
dtree = model.fit(X_train,Y_train)
```

```python
#Score the model

score = model.score(X_test, Y_test)
print("High-Dimensionality Score: ", round((score*100), 3))
```

```python
# Install GraphViz
# This code will create a file that can be pasted into graphviz to see the tree

tree.export_graphviz(model.tree_, out_file = 'tree.dot', feature_names = X.columns)
```
